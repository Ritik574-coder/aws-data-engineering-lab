```python
import boto3
import pprint
from botocore.exceptions import ClientError 
s3 = boto3.client('s3')
```

```python
def create_s3_bucket(bucket_name : str , region="ap-south-1") -> None : 
    try :
        s3.create_bucket(
            Bucket=bucket_name,
            CreateBucketConfiguration={
                "LocationConstraint" : region
            }
        )

        print(f"Bucket {bucket_name} successfully created ")

    except Exception as error :
        print("Faild to create s3 bucket")
        raise error
```

```python
create_s3_bucket("ritsky-demo-bucket-4044")
```

```python
def list_all_bucket() :
    bucket_list = s3.list_buckets()

    print("THE LIST OF ALL BUCKET NAME : ")
    for bucket in bucket_list["Buckets"] :
        print( "    Bucket Name : ", bucket['Name'])
```

```python
list_all_bucket()
```

```python
def check_bucket_exist(bucket_name) -> None : 
    try : 
        s3.head_bucket(Bucket=bucket_name)
        return True 
    except ClientError : 
        return False

for bucket in [
    "ritsky-demo-bucket-4044",
    "nyc-yellow-taxi-dataset-2026",
    "ritsky-bucket-2025",
    "my_new_demo_bucket"
]: 
    print(f"{bucket} : {check_bucket_exist(bucket)}")
```

```python
def delete_bucket(bucket_name) -> None :
    """
        Theis function are build for deleting the aws s3 bucket and it have one parameter bucket_name 
        in bucket name write the bucket name that you want to delete 
        
        INFO : THIS FUNCTION ARE ONLY BUILD FOR DELETING EMPTY BUCKET SO MAKE SOURE YOU BUCKET ARE EMPTY 
    """

    try : 
        s3.delete_bucket(Bucket=bucket_name)
        print(f"Bucket {bucket_name} successfully deleted ")

    except ClientError :
        print("An error during bucket deleting")
        raise ClientError 
```

```python
delete_bucket("ritsky-demo-bucket-4044")
```

```python
response = s3.list_objects_v2(Bucket="ritsky-bucket-2025")

for obj in response.get("Contents", []):
    print(obj["Key"])
```

```python
response = s3.list_objects_v2(
    Bucket="ritsky-bucket-2025",
    Delimiter="/"
)

for prefix in response.get("CommonPrefixes", []):
    print(prefix["Prefix"])
```
